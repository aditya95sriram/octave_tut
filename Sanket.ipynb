### To plot the graph of a function, you need to take the following steps −

   1. Define x, by specifying the range of values for the variable x, for which the function is to be plotted

   2. Define the function, y = f(x)

   3. Call the plot command, as plot(x, y)

#### To plot the example vectors above in a new figure:

   

```matlab
clear all % clear all previous variables
X = [3 9 27]; % my dependent vector of interest
t = [1 2 3]; % my independent vector
figure % create new figure
plot(t, X)
    
```

```matlab
clear all
x = [1 2 3 4 5 6 7 8 9]; %Very discrete points 
y = sin(x);
plot(y,x);
```

```matlab
clear all;
x = linspace(0,10); % generates a row vector of 100 linearly equally spaced points between X1 and X2.
y = sin(x);
plot(y,x);
```

## Labeling Axes:

```matlab
plot(x,y);
title('Plot of Distance over Time') % title
ylabel('Distance (m)') % label for y axis
xlabel('Time (s)') % label for x axis
```

### Plotting different functions in one figure and distinguish them from each other via a legend

```matlab
figure;

x = linspace(-2,2,4000);

y = (sqrt(cos(x)).*cos(200*x)+sqrt(abs(x))-0.7).*(4-x.*x).^0.01; % plot Heart Shape

f = -exp(x)/10;  % plot some exponential function 

z = sin(x);  % plot sin function
plot(x, y, 'red', x, f, 'blue', x, z, 'green'); % plot( x, y, colour)

legend('Heart Shape','exponential function',' sin function')
axis([-2,5,-2,3]);
```

## Subplots

It can sometimes be useful to display multiple plots on the same figure for comparison.
This can be done using the subplot function, that takes arguments for number of rows of plots,
number of columns of plots, and plot number currently being plotted:

```matlab
clear all
close all
% subplot (nrows,ncols,plot_number)
x=0:.1:2*pi; % x vector from 0 to 2*pi, dx = 0.1
subplot(2,2,1); % plot sine function
plot(x,sin(x));
subplot(2,2,2); % plot cosine function
plot(x,cos(x));
subplot(2,2,3) % plot negative exponential function
plot(x,exp(-x));
subplot(2,2,4); % plot x^3
plot(x, x.^3);
```

## 3D

```matlab
clear all
close all
[x,y] = meshgrid([-2:.1:2]); % set up 2-D plane
Z = x.*exp(-x.^2-y.^2); % plot 3rd dimension on plane
figure
surf(x, y, Z) % surface plot, with gradient(Z)
% determining color distribution
colorbar % display color scale, can adjust
% location similarly to legend
```

```matlab
clear all
close all
x = linspace(-2,2,70);
y = x;
[x,y] = meshgrid(x,y); % set up 2-D plane
Z = x.*exp(-x.^2-y.^2); % plot 3rd dimension on plane
figure
mesh(x,y,Z); % surface plot, with gradient(Z)
% determining color distribution
colorbar % display color scale, can adjust
```

```matlab
clear all
close all
x = linspace(-2,2,500);
y = x;
[x,y] = meshgrid(x,y); % set up 2-D plane
Z = x.*exp(-x.^2-y.^2); % plot 3rd dimension on plane
figure
surfl(x,y,Z); % surface plot, with gradient(Z)
% determining color distribution
colorbar % display color scale, can adjust
% location similarly to legend
colormap(cool)
shading interp %The shading function controls the color shading of surface 
camlight right %Create or move light object in camera coordinates
lighting gouraud %Specify lighting algorithm
```

```matlab
n = 51;
x=linspace(-5., 5., n);
y=linspace(-5., 5., n);
[x, y]=meshgrid(x, y);
sinc = @(x) sin(x)./x;
z=sinc(x.*y);
z(isnan(z)) = 1;

nn = 401;
xi = linspace(-5.0, 5.0, nn);
yi = xi;
[xi, yi] = meshgrid(xi, yi);
zi = interp2(x, y, z, xi, yi, 'spline');
surf(xi, yi, zi, 'LineStyle', 'none', 'FaceColor', 'interp')
colormap(cool)
```

```matlab
u = linspace(0,2*pi,20); v = linspace(-1,1,20); % domain parameterization
[U,V] = meshgrid(u,v); % creating 2-D arrays
% Use of the 2-D arrays for component-wise operations
X = (2+V.*cos(U/2)).*cos(U); % x(u,v)
Y = (2+V.*cos(U/2)).*sin(U); % y(u,v)
Z = V.*sin(U/2); % z(u,v)
% Plotting (3-D graph)
figure; % opening new window for Figure 4
surfl(X,Y,Z); % 3-D graph plotting (Figure 4)
colormap(cool)
shading interp
lighting flat
title('Mobius strip');
axis([-3 3 -3 3 -1 1]);
xlabel('x'); ylabel('y'); zlabel('z');
```

### Symbolic computation
We can do Symbolic Computation in MATLAB. "Symbolic Computation" is the computation with symbolic expressions.


```matlab
clear all
close all
syms x % Symbolic Expression
a = sym(2) % Symbolic Variable
x = sqrt(a)

y = 2
z = sqrt(y) % Compare symbolic expression and this expression
```

Expansion and Factorization of Equation

```matlab
syms  x  
p=(x-1)*(x-2)*(x-3)  
p =
(x - 1)*(x - 2)*(x - 3)  

expand(p)  
ans =
x^3 - 6*x^2 + 11*x - 6  

factor(ans)  
ans =
(x - 3)*(x - 1)*(x - 2)  
```

Differentiating function

```matlab
syms x  
f=sin(x^2)  
f =
sin(x^2)  

diff(f,x)  
ans =
2*x*cos(x^2)  
```
